```python
#불 자료형은 어떻게 사용할까? (True, False)
a = True
b = False
print(a)
print(b)

print(type(a))
print(type(b))
```

```python
print(1 == 1)
print(2 > 1)
print(2 < 1)
print(2 >= 2)
print(2 <= 2)
print(2 != 2) # != 다르다
```

```python
# if 조건문 :
	# 수행할 문장
#bool은 excel if 조건문 에서 조건문(참,거짓)에 해당 True, False 인지 판단

print(bool("python"))
print(bool("")) #문자열에선 "" 제외하곤 전부다 True
print(bool(" "))

print(bool(10)) #숫자에서는 0을 제외하고 전부다 True
print(bool(0))

print(bool([1, 2, 3]))
print(bool([])) #리스트 빈칸은 False

print(bool({"name" : "pey"}))
print(bool({})) #딕셔너리 빈칸 False
```

```python
#논리연산자

## and 연산자
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(True and True and False)
```

```python
# or 연산자
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print(True or True or False)
```

```python
# not 연산자
print(not True)
```
