```python
# 문자열 포매팅 
## 1. 숫자 바로 대입(%d)
"I eat %d apples" % 3
print("I eat %d apples" % 3)

## 2. 문자열 바로 대입(%s)
print("I eat %s apples" % "five")
```

```python
## 3. 숫자값을 나타내는 변수로 대입
number = 3
print("I eat %d apples" % number)
```

```python
## 2개 이상의 값 넣기

number = 10
day = "three"
print("I eat %d apples so I was sick for %s days" % (number, day))
```

```python
# format 함수를 사용한 포매팅 %d, %s 없이 바로 사용가능

print("I eat {0} apples".format(3))
print("I eat {0} apples".format("five"))
```

```python
# format함수를 이용해서 2개 이상의 값 넣기
number = 10
day = "three"
print("I eat {0} apples so I was sick for {1} days".format(number, day))
print("I eat {1} apples so I was sick for {0} days".format(number, day))
```

```python
# f문자열 포매팅 (가장 많이 사용하는 포매팅 / 큰 따옴표 앞에 f를 적어주면 된다)

name = "홍길동"
age = 10
print("나의 이름은 {name}입니다. 나이는 {age}입니다.")
print(f"나의 이름은 {name}입니다. 나이는 {age}입니다.")
```

```python
# 문자열에 관련 함수들 

a = "hobby" #a라는 문자열을 가진 객채 뒤에 .을 붙여 함수의 사용 가능함
print(type(a))

## 문자 개수 세기 - count / lan 과 구분하여 사용할 것 
print(a.count("b"))
```

```python
## 위치 알려주기 - find / 있으면 해당하는 인덱스를 출력, 없으면 -1을 출력

a = "Python is the best choice"
print(a.find("b"))
print(a.find("k"))
```

```python
# 위치 알려주기2- index
a = "Life is too short"
print(a.index("t")) #이 경우 좌로부터 첫번째 만 찾음
print(a.index("k"))
```

```python
# 문자열  삽입 함수 - join

print(",".join("abcd"))
```

```python
# 소문자를 대문자로 바꾸기 - upper
a = "hi"
print(a.upper())
```

```python
# 대문자를 소문자로 바꾸기 - lower
a = "HI"
print(a.lower())
```

```python
# 공백지우기 - strip
a = "   hi   "
print(a)
a= a.strip()
print(a)
```

```python
# 공백지우기 왼쪽(l+strip) / 오른쪽(r+strip)
a = "   hi   "
print(a.lstrip())
print(a.rstrip())
```

```python
# 문자열 바꾸기 - replace
a = "Life is too short"
a.replace("Life", "your leg")
```

```python
# 문자열 나누기 - split (나누었을 때 결과값도 확인할 것 / 공백을 기준으로 나눔? 기호를 입력하면 입력된 기호를 기준으로 분리)
a = "Life is too short"
a.split()
```
