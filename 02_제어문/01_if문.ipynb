```python
# if문은 왜 필요할까?
## 돈이 있으면 택시를 타고 가고, 돈이 없으면 걸어가라

money = True
if money == True:
	print("택시를 타고 가라")
else:
	print("걸어가라")
```

```python
"만약 3000원 이상의 돈을 가지고 있으면 택시를 타고 가고 그렇지 않으면 걸어가라"
money = 3000
if money >= 3000:
	print("택시를 타고가라")
else:
	print("걸어가라")
```

```python
# "돈이 3000원 이상이 있거나 카드가 있다면 택시를 타고 가고 그렇지 않으면 걸어가라"

money = 2000
card = True
if money >= 3000 or card == True:
	print("택시를 타고가라")
else:
	print("걸어가라")
```

```python
money = 2000
card = True
if (money >= 3000) or (card == True):
	print("택시를 타고가라")
else:
	print("걸어가라")
```

```python
#다양한 조건을 판단하는 elif
# 주머니에 돈이 있으면 택시를 타고가고
# 주머니에 돈은 없지만 카드가 있으면 택시를 타고가고
# 돈도 없고 카드도 없으면 걸어가라

pocket = ["paper", "phone"]
card = False

if "money" in pocket:
	print("1택시를 타고 가라")
elif card == True:
	print("2택시를 타고 가라")
else:
	print("걸어가라")
```
