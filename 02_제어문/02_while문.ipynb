```python
coffee = 5
money = 300

while money:
	print("돈을 받았으니 커피를 줍니다")
	coffee = coffee - 1
	print(f"남은 커피는 {coffee}개 입니다")

	if coffee == 0:
		print("커피가 다 떨어졌습니다.")
		break
```

```python
coffee = 5
while True:
	money = int(input("돈을 넣어주세요: "))
	if money == 300:
		print("커피를 줍니다")
		coffee -= 1 # coffee = coffee -1
	elif money > 300:
		print(f"거스름돈 {money - 300}원을 주고 커피를 줍니다")
		coffee -= 1
	else:
		print("돈을 다시 돌려주고 커피를 주지 않습니다.")
		print(f"현재 남은 커피는 {coffee}개 입니다.")

	if coffee == 0:
		print("커피가 품절 되었습니다.")
		break
```

```python
# while 문의 맨 처음으로 돌아가기 - continue(다음이아닌 while로 돌아가는 것) / break만나면 종료

a = 0
while a < 10:
	a = a + 1
	if a % 2 == 0:
		continue
	print(a)
```
