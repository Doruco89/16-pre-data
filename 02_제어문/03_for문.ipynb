```python
test_list = ["one", "two", "three"] #내용이 없으면 거기서 종료

for tl in test_list:
	print(tl)
```

```python
a = [(1, 2), (3, 4), (5, 6)] #딕셔너리를 items를 이용하여 풀어헤쳤을때의 모습 / 구조분해, unpacking(python에서)

for k, j in a:
	print(k, j)
```

```python
marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
	number = number + 1
	if mark >= 60:
		print(f"{number} 학생은 합격입니다.")
	else:
		print(f"{number} 학생은 불합격입니다.")
```

```python
# for문과 continue 
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
	number += 1
	if mark < 60: continue

	print(f"{number}학생은 합격입니다.")
```

```python
# for와 range 를 이용한 구구단
for i in range(7, 10):
	for j in range(1, 9+1):
		print(f"{i} X {j} = {i * j}")
```

```python
# 리스트 컴프리헨션 사용하기 (축약어, 굉장히 많이 사용)

a = [1, 2, 3, 4[]
result = []

for num in a:
	result.append(num * 3)

print(result)
```

```python
a = [1, 2, 3, 4]
result = [num * 3 for num in a ]
print(result)
```

```python
# for문과 break문
for i in range(10): # range(0,10) range는 함수이므로 범위불가 인수로
	if i == 5:
		break
	print(i)
```

```python
# enumerate() 함수 사용빈도 높음 / 사용시 인덱스 번호를 알려줌
fruits = ['apple', 'banana', 'orange']
for fruit in fruits:
	print(f"과일: {fruit}")
```

```python
fruits = ['apple', 'banana', 'orange']
for i, fruit in enumerate(fruits):
	print(f"{i+1}번째 과일: {fruit}")
```

```python
# zip함수로 여러 리스트 순회하기 (같은 인덱스 순서끼리 튜플로 묶음)
names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]

for item in zip(names, scores):
	print(item)
```

```python
names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]

for name, score in zip(names, scores):
	print(f"{name}: {score}점")
```
