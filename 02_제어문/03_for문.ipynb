```python
# for문의 기본 구조 (if, for이 가장 많이 쓰임)
# for 변수 in 리스트(또는 튜플, 문자열):
# 	수행할 문장1
# 	수행할 문장2
```

```python
test_list = ["one", "two", "three"] #내용이 없으면 거기서 종료

for tl in test_list:
	print(tl)
```

```python
a = [(1, 2), (3, 4), (5, 6)] #딕셔너리를 items를 이용하여 풀어헤쳤을때의 모습 / 구조분해, unpacking(python에서)

for k, j in a:
	print(k, j)
```

```python
marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
	number = number + 1
	if mark >= 60:
		print(f"{number} 학생은 합격입니다.")
	else:
		print(f"{number} 학생은 불합격입니다.")
```

```python
# for문과 continue 
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
	number += 1
	if mark < 60: continue

	print(f"{number}학생은 합격입니다.")
```

```python
# for와 range 를 이용한 구구단
for i in range(2, 10):
	for j in range(1, 9+1):
		print(f"{i} X {j} = {i * j}")
```
