```python
# 리스트는 어떻게 만들고 사용할까? (하나의 변수에 여러가지 요소가 들어간 것)

# 리스트명 = [요소1, 요소2, ...]

a = []
b = [1, 2, 3]
c = ["Life, is, too, short"]
d = [1, 2, "Life", "is"] #숫자, 문자열 복합하여 리스트에 포함 가능
e = [1, 2, ["Life", "is"]] #리스트 안에 리스트를 포함할 수 있음

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
# 리스트의 인덱싱과 슬라이싱
a = [1, 2, 3]
print(a[0]) #문자열과 마찬가지로 첫번째는 0, 출력된 값은 숫자이다.
print(a[2])
print(type(a[2]))
```

```python
a[0] + a[2]
```

```python
a[-1]
```

```python
b = ['1', '2', '3', 'a', 'b', 'c']
b[-1]
```

```python
b[0] + b[2] + b[-1]
```

```python
c = [1, 2, ['a', 'b', ['Life', 'is']]]
len(c) #요소는 3가지
c[-1][-1][0]
```

```python
# 리스트의 슬라이싱
a = [1, 2, 3, 4, 5]

print(a[0:1])
print(a[0:2])
print(a[3:3])
```

```python
print(a[2:])
print(a[:3])
```

```python
a = [1, 2, 3, ['a', 'b', 'c'], 4, 5]

# ['b', 'c']
len(a)
```

```python
a[3][1:2] #리스트 형태의 데이터를 슬라이싱 해왔으므로 결과값이 리스트 형태로 출력
```

```python
# 리스트 더하기 (+)
a = [1, 2, 3]
b = [4, 5, 6]

a + b
```

```python
# 리스트 곱하기 (*)
a = [1, 2, 3]
a * 3
```

```python
# 리스트의 수정과 삭제

a = [1, 2, 3]
a[2] = 4
print(a)
```

```python
# 리스트의 요소 삭제하기 - del
a = [1, 2, 3]
del a[1]
print(a)
```

```python
# 슬라이싱을 이용하여 범위로 삭제?
a =[1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
# 리스트 관련 함수
## 리스트 요소 추가하기 - append / 사용빈도 높음

a = [1, 2, 3]
a.append(4)
print(a)
```

```python
a.append([5, 6])
print(a)
```

```python
# 리스트 정렬 - sort (내장함수에도 sorted도 있음)
a = [1, 4, 2, 3]
a.sort()
print(a)
a.sort(reverse=True)
print(a)
```

```python
a = [1, 4, 2, 3]
sorted(a)
sorted(a, reverse=True)
```

```python
# 리스트 뒤집기 - reverse 기준없이 그냥 방향의 전환
a = ['a', 'c', 'b']

a.reverse()
print(a)
```

```python
# 인덱스 반환 - index
a = ['a', 'c', 'b']
a.index("c")
```

```python
# 리스트의 요소 삽입 - insert (append와 차이점을 볼 것, append는 마지막에 추가?)
a= [1, 2, 3]
a.insert(0, 4)
print(a)
```

```python
# 리스트의 요소 제거 - remove (del은 명령어 다른개념)
a = [1, 2, 3, 1, 2, 3]

a.remove(3)
a.remove(3)
print(a)
```

```python
# 리스트의 요소 꺼내기 - pop (빅데이터 분석기사에 주로 사용 / 교육과정과 기사과정이 비슷하므로 기사 준비해볼 것 추천)
a = [1, 2, 3]
result = a.pop() #공백시 기본값 -1이므로 가장 뒷요소인 3을 추출
print(result)
print(a)
```

```python
# 리스트의 포함된 요소 x의 개수 세기 - count
a = [1, 2, 3, 1]
a.count(1)
```

```python
# 리스트 확장 - extend / +연산과 비슷한 개념으로 이해

a = [1, 2, 3]
a.extend([4, 5])
print(a)
```
