```python
# 딕셔너리는 어떻게 만들까? /key와 value가 하나의 쌍으로 이루어진
#{key1 : value1, key2 : value2, ...}

dic = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(dic)
print(type(dic))
```

```python
a = 1
a = 2
print(a) # 재할당 되는 경우 올 수 있음. (숫자, 문자, 튜플 은 올 수 있다.)
```

```python
# 딕셔너리에 key 값으로 올수 없는 데이터 타입은 리스트 -> mutable
# 리스트는 변경 가능하기 때문에 key 값으로 올 수 없다.
a = [1, 2, 3]
a[0] = 3
print(a)
```

```python
# 딕셔너리 쌍 추가하기
a = {1 : "a"}
a[2] = 'b'
#2는 인덱스 번호가 아니라 dic안에 key값 / 문자열,리스트,튜플과 다른점
print(a)
```

```python
a[3] = [1, 2, 3]
print(a)
```

```python
# 딕셔너리 요소 삭제하기 del 명령어 사용 (key 값을 삭제)
print(a)
del a[1]
print(a)
```

```python
# 딕셔너리 key를 사용해서 value 얻기
grade = {
	'pey' : 10,
	'juli' : 99
}

print(grade['pey'])
print(grade['juli'])
```

```python
# 딕셔너리 관련 함수(List의 .uppend 같은)
## 1. key 리스트 만들기 - keys

a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
a.keys() 
#리스트 화 전 준비단계
list(a.keys()) #리스트 화
```

```python
## 2. value 리스트 만들기 - values
a.values()
list(a.values())
```

```python
## 3. key, value 쌍으로 얻기 - items
print(a)
a.items() #꺼내 쓸 수 있는 상태
list(a.items()) #하나의 요소는 튜플 형태로 존재
```

```python
## 4. key, value 쌍 모두 지우기 - clear
a.clear()
print(a)
```

```python
# key로 value 얻기 - get
a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(a['name'])
print(a.get('name')) #get 사용
```

```python
print(a['hobby'])
a.get('hobby',0) #에러처리에 도움이된다.
```

```python
# 해당 key가 딕셔너리안에 있는지 조사하기 - in
print(a)
print('name' in a)
print('email' in a)
```

```python
# key로 value 꺼내기 - pop(뽑아낸다?)
a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
result = a.pop('phone')
print(result)
print(a)
```
